# multiple seasons

```python
leagues = ['E0','E2','E3']
frames = []

for league in leagues:
    for season in range(15,21):
        df = pd.read_csv(root + (str(season) + str(season+1)) + '/' + league + '.csv',encoding = 'unicode_escape') # if encoding error comes put encoding = 'unicode_escape'
        df.insert(1,'season',season)
        frames.append(df)
```

# Organize All The Data in a Dictionary

```python
# create a dictionary with original name of league as key
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1', 
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
dict_historical_data = {} 
frames = []

for league in dict_countries:
    frames = []
    for season in range(15,21):
        df = pd.read_csv(root + (str(season) + str(season+1)) + '/' + dict_countries[league] + '.csv',encoding = 'unicode_escape') # if encoding error comes put encoding = 'unicode_escape'
        df.insert(1,'season',season)
        df.insert(1,'league_name',league)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat 
```
