### Conditional column from more than 2 Choices: np.select()

```python
# create a array based on the multiple choices you want to give ( in np.where() method we only have two chioces )
# create two listst one conditions and other is the values
conditions = [
    df_laptops['Price_euros']>3000,
    (df_laptops['Price_euros']>2000) & (df_laptops['Price_euros']<=3000),
    (df_laptops['Price_euros']<=2000) & (df_laptops['Price_euros'] >1000),
    df_laptops['Price_euros']<=1000
]
values = ['Too Exapensive','Expensive','Affordable','Cheap']
```

```python
df_laptops['Price tiers'] = np.select(conditions,values,default = 'Cheap123')
```
